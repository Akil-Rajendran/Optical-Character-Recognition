## Display Tamil Image

```
# show an image
import PIL
from PIL import ImageDraw
im = PIL.Image.open("img.png")
im
```

### Download Tamil and English Models 

```
# Create a reader to do OCR.
import easyocr
reader = easyocr.Reader(['ta','en'])
```

## Perform OCR and extract Bounding Boxes

```
# Doing OCR. Get bounding boxes.
bounds = reader.readtext('img.png')
bounds
```

### Draw the extracted Bounding boxes

```
# Draw bounding boxes
def draw_boxes(image, bounds, color='blue', width=2):
    draw = ImageDraw.Draw(image)
    for bound in bounds:
        p0, p1, p2, p3 = bound[0]
        draw.line([*p0, *p1, *p2, *p3, *p0], fill=color, width=width)
    return image

draw_boxes(im, bounds)
```
